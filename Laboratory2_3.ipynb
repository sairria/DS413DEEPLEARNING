```python
import numpy as np
```

```python
#1 Define Activation Functions & their derivatives

def relu(z):
    """ReLU activation function: max(0, z)"""
    return np.maximum(0, z)

def relu_derivative(z):
    """Derivative of ReLU: 1 if z > 0, 0 otherwise"""
    # Create a copy of z and set values based on the condition
    dz = np.where(z > 0, 1, 0)
    return dz
```

```python
#2 Initialize Parameters and Inputs from Laboratory Task 2 ---

# Input vector (x) and True output (y)
# x = np.array([1, 0, 1])
# y = np.array([1])
x = np.array([[1], [0], [1]]) # Input is a 3x1 column vector
y_true = np.array([[1]])      # True output is a 1x1 vector

# Hidden unit weights (W1: 3x2 matrix)
W1 = np.array([
    [0.2, -0.3],
    [0.4, 0.1],
    [-0.5, 0.2]
])

# Output unit weights (W2: 2x1 matrix)
# Note: The problem shows W21 and W22 in a column, but for matrix multiplication
# from hidden layer (2x1) to output (1x1), W2 should be 1x2 or a 2x1 vector.
# Assuming hidden layer has 2 units, and output has 1 unit.
# The shape of W2 should be 2x1 if the hidden activation is 2x1.
W2 = np.array([
    [-0.3],
    [-0.2]
])

# Biases (Thetas)
# Theta_1 and Theta_2 are for the two hidden units, Theta_3 for the output unit.
theta_1 = np.array([[-0.4], [0.2]])  # Biases for the 2 hidden units (2x1)
theta_2 = np.array([[0.1]])          # Bias for the 1 output unit (1x1)

# Learning rate
lr = 0.001

print(f"Initial W1:\n{W1}\n")
print(f"Initial W2:\n{W2}\n")
```

```python
# Input vector (x) and True output (y)
# x = np.array([1, 0, 1])
# y = np.array([1])
x = np.array([[1], [0], [1]]) # Input is a 3x1 column vector
y_true = np.array([[1]])      # True output is a 1x1 vector

# Hidden unit weights (W1: 3x2 matrix)
W1 = np.array([
    [0.2, -0.3],
    [0.4, 0.1],
    [-0.5, 0.2]
])

# Output unit weights (W2: 2x1 matrix)
# Note: The problem shows W21 and W22 in a column, but for matrix multiplication
# from hidden layer (2x1) to output (1x1), W2 should be 1x2 or a 2x1 vector.
# Assuming hidden layer has 2 units, and output has 1 unit.
# The shape of W2 should be 2x1 if the hidden activation is 2x1.
W2 = np.array([
    [-0.3],
    [-0.2]
])

# Biases (Thetas)
# Theta_1 and Theta_2 are for the two hidden units, Theta_3 for the output unit.
theta_1 = np.array([[-0.4], [0.2]])  # Biases for the 2 hidden units (2x1)
theta_2 = np.array([[0.1]])          # Bias for the 1 output unit (1x1)

# Learning rate
lr = 0.001

print(f"Initial W1:\n{W1}\n")
print(f"Initial W2:\n{W2}\n")
```

```python
#3 Forward Propagation 

print("--- FORWARD PASS ---")

# Step 1: Calculate the weighted input for the hidden layer (Z1)
# Z1 = W1^T * x + theta_1. Since W1 is 3x2 and x is 3x1, we need to transpose W1.Alternatively, use W1 (3x2) and transpose x (1x3), then transpose the result. A cleaner way is Z1 = x^T @ W1 + theta_1^T, then transpose Z1 for a 2x1 vector. Let's stick to the convention: Z1 = W1_T @ x + b1. Here, W1 is (Input_Size x Hidden_Size). We assume x is a 3x1 vector. Z1 will be (Hidden_Size x 1) = (2x1)
Z1 = np.dot(W1.T, x) + theta_1
# Z1 = [[0.2 * 1 + 0.4 * 0 + (-0.5) * 1 - 0.4], [-0.3 * 1 + 0.1 * 0 + 0.2 * 1 + 0.2]]
print(f"Z1 (Hidden Weighted Input):\n{Z1}")
# Expected Z1: [[-0.7], [0.1]]

# Step 2: Apply ReLU activation to get hidden layer output (A1)
A1 = relu(Z1)
print(f"A1 (Hidden Activation):\n{A1}")
# Expected A1: [[0.0], [0.1]]

# Step 3: Calculate the weighted input for the output layer (Z2)
# Z2 = W2^T * A1 + theta_2. W2 is 2x1, A1 is 2x1. Z2 will be (Output_Size x 1) = (1x1)
Z2 = np.dot(W2.T, A1) + theta_2
# Z2 = [[-0.3 * 0.0 + (-0.2) * 0.1 + 0.1]]
print(f"Z2 (Output Weighted Input):\n{Z2}")
# Expected Z2: [[0.08]
```

```python
# Step 4 Apply ReLU activation to get final prediction (y_pred)
# For the output layer, an identity function is sometimes used, but the prompt specifies 'relu' as *the* activation function.
y_pred = relu(Z2)
print(f"y_pred (Final Prediction):\n{y_pred}")
# Expected y_pred: [[0.08]]
```

```python
# Step 5: Calculate the Loss (Mean Squared Error)
loss = np.sum((y_pred - y_true)**2) / 2 # MSE simplified for single output
print(f"Loss (E):\n{loss}")
# Expected Loss: (0.08 - 1)^2 / 2 = 0.8464 / 2 = 0.4232
```

```python
#4. Backward Propagation (Gradient Calculation)
print("\n--- BACKWARD PASS (Gradient Calculation) ---")
```

```python
# 1. Gradient of Loss w.r.t. Predicted Output (dE/dy_pred)
dE_dy_pred = y_pred - y_true # d/dy_pred [1/2 * (y_pred - y_true)^2]
print(f"dE_dy_pred:\n{dE_dy_pred}")
# Expected: [[0.08 - 1]] = [[-0.92]] 
```

```python
# 2. Gradient of Predicted Output w.r.t. Z2 (dy_pred/dZ2)
dy_pred_dZ2 = relu_derivative(Z2)
print(f"dy_pred_dZ2:\n{dy_pred_dZ2}")
# Expected: [[1]] since Z2 = 0.08 > 0
```

```python
# 3. Gradient of Loss w.r.t. Z2 (dE/dZ2) -> The 'delta' for output layer
# Chain rule: dE/dZ2 = (dE/dy_pred) * (dy_pred/dZ2)
dE_dZ2 = dE_dy_pred * dy_pred_dZ2
print(f"dE_dZ2 (Output Delta):\n{dE_dZ2}")
# Expected: [[-0.92]] * [[1]] = [[-0.92]]
```

```python
# 4. Gradient of Loss w.r.t. W2 (dE/dW2)
# dE/dW2 = dE/dZ2 * A1^T
dE_dW2 = np.dot(A1, dE_dZ2.T) # A1 is 2x1, dE_dZ2.T is 1x1. Result is 2x1.
print(f"dE_dW2:\n{dE_dW2}")
# Expected: [[0.0], [0.1]] * [[-0.92]] = [[0.0], [-0.092]]
```

```python
# 5. Gradient of Loss w.r.t. Bias Theta_2 (dE/d_theta_2)
# dE/d_theta_2 = dE/dZ2
dE_dtheta_2 = dE_dZ2
print(f"dE_dtheta_2:\n{dE_dtheta_2}")
# Expected: [[-0.92]]
```

```python
#B. Hidden Layer Gradients (Layer 1)
```

```python
# 6. Gradient of Loss w.r.t. Hidden Activation (dE/dA1)
# Chain rule: dE/dA1 = W2 * dE/dZ2. W2 is 2x1, dE/dZ2 is 1x1. No, W2 is 2x1, dE/dZ2 is 1x1
# dE/dA1 = W2 * dE/dZ2.T, W2 is 2x1, dE/dZ2.T is 1x1. Result 2x1.
dE_dA1 = np.dot(W2, dE_dZ2) # W2 (2x1) * dE_dZ2 (1x1). Result is 2x1.
print(f"dE_dA1:\n{dE_dA1}")
# Expected: [[-0.3], [-0.2]] * [-0.92] = [[0.276], [0.184]]
```

```python
# 7. Gradient of Hidden Activation w.r.t. Z1 (dA1/dZ1)
dA1_dZ1 = relu_derivative(Z1)
print(f"dA1_dZ1:\n{dA1_dZ1}")
# Expected: [[0], [1]] since Z1 = [[-0.7], [0.1]]
```

```python
# 8. Gradient of Loss w.r.t. Z1 (dE/dZ1) -> The 'delta' for hidden layer
# Chain rule: dE/dZ1 = (dE/dA1) * (dA1/dZ1) (element-wise multiplication)
dE_dZ1 = dE_dA1 * dA1_dZ1
print(f"dE_dZ1 (Hidden Delta):\n{dE_dZ1}")
# Expected: [[0.276 * 0], [0.184 * 1]] = [[0.0], [0.184]]
```

```python
# 9. Gradient of Loss w.r.t. W1 (dE/dW1)
# dE/dW1 = x * dE/dZ1^T. x is 3x1, dE_dZ1.T is 1x2. Result is 3x2.
dE_dW1 = np.dot(x, dE_dZ1.T)
print(f"dE_dW1:\n{dE_dW1}")
# Expected:
# [1] * [0.0, 0.184] = [0.0, 0.184]
# [0] * [0.0, 0.184] = [0.0, 0.0]
# [1] * [0.0, 0.184] = [0.0, 0.184]
# Result: [[0.0, 0.184], [0.0, 0.0], [0.0, 0.184]]
```

```python
# 10. Gradient of Loss w.r.t. Bias Theta_1 (dE/d_theta_1)
# dE/d_theta_1 = dE/dZ1
dE_dtheta_1 = dE_dZ1
print(f"dE_dtheta_1:\n{dE_dtheta_1}")
# Expected: [[0.0], [0.184]]
```

```python
# 5. Update Parameters 

print("\n--- PARAMETER UPDATE ---")

# W2 Update
W2_new = W2 - lr * dE_dW2
print(f"W2_new:\n{W2_new}")
# W2[1] = -0.2 - 0.001 * (-0.092) = -0.2 + 0.000092 = -0.199908

# Theta_2 Update
theta_2_new = theta_2 - lr * dE_dtheta_2
print(f"theta_2_new:\n{theta_2_new}")
# theta_2 = 0.1 - 0.001 * (-0.92) = 0.1 + 0.00092 = 0.10092

# W1 Update
W1_new = W1 - lr * dE_dW1
print(f"W1_new:\n{W1_new}")
# W1[0, 1] = -0.3 - 0.001 * 0.184 = -0.3 - 0.000184 = -0.300184
# W1[2, 1] = 0.2 - 0.001 * 0.184 = 0.2 - 0.000184 = 0.199816

# Theta_1 Update
theta_1_new = theta_1 - lr * dE_dtheta_1
print(f"theta_1_new:\n{theta_1_new}")
# theta_1[1] = 0.2 - 0.001 * 0.184 = 0.2 - 0.000184 = 0.199816
```
